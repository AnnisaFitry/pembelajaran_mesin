## Preprocessing

```python
import pandas as pd
import numpy as np
df = pd.read_excel('data/dataset_quiz.xlsx')
df.head(100)
```

### Drop kolom yang tidak berpengaruh

```python
# dalam data tersebut yaitu pada kolom keterlibatan_vltr, luaran_hki, dan luaran_lain dijelaskan lebih rinci pada kolom sesudahnya.
# jadi kolom tersebut tidak terlalu berpengaruh.
df.drop('keterlibatan_vltr', axis=1, inplace=True)
df.drop('luaran_hki', axis=1, inplace=True)
df.drop('luaran_lain', axis=1, inplace=True)
df
```

### Mengecek Missing Value

```python
# Dari data di atas, telah ditemukan beberapa nilai kosong atau missing value.
# Untuk pengecekan secara mendetail yaitu dengan :
print(f'Jumlah data missing value : \n{df.isnull().sum()}\n\n')
print(f'Jumlah data yang tidak kosong : \n{df.notna().sum()}')
```

### Mengisi Missing Value dengan Mean

```python
# Diketahui hasil di atas menunjukkan bahwa di setiap baris dan kolom tidak semuanya terdapat missing value.
# Missing value terjadi pada kolom id_ketua sampai id_anggota5, ini dikarenakan penghapusan anggota yang tidak aktif.
# Missing value ini dapat disebut dengan hilang secara tidak acak atau Missing Not At Random (MNAR).
# Jadi, menurut saya hal ini kemungkinan akan mempengaruhi data.
# Akan lebih baik data yang hilang dilakukan Imputation.

# Metode Data Frame : fillna() dengan mean() yaitu mengisi data kosong dengan rata-rata data kolom.
df['id_ketua'].fillna(df['id_ketua'].mean(), inplace=True)
df['id_anggota1'].fillna(df['id_anggota1'].mean(), inplace=True)
df['id_anggota2'].fillna(df['id_anggota2'].mean(), inplace=True)
df['id_anggota3'].fillna(df['id_anggota3'].mean(), inplace=True)
df['id_anggota4'].fillna(df['id_anggota4'].mean(), inplace=True)
df['id_anggota5'].fillna(df['id_anggota5'].mean(), inplace=True)
df.head()
```

```python
print(f'Info : \n{df.info()}\n\n')
print(f'Deskrisi : \n{df.describe()}')
```

## Metode Pembagian Data

### Random Split

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:1:1
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0)

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)

# Cek masing-masing ukuran data

print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data val: {df_val.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

# Cek rasio tiap label
# Hak Cipta
print('=================================')
print(f'Jumlah label data asli:\n{df.jenis_luaran_hki.value_counts()}')
print(f'Jumlah label data train:\n{df_train.jenis_luaran_hki.value_counts()}')
print(f'Jumlah label data val:\n{df_val.jenis_luaran_hki.value_counts()}')
print(f'Jumlah label data test:\n{df_test.jenis_luaran_hki.value_counts()}')
# Publikasi
print('=================================')
print(f'Jumlah label data asli:\n{df.jenis_luaran_lain.value_counts()}')
print(f'Jumlah label data train:\n{df_train.jenis_luaran_lain.value_counts()}')
print(f'Jumlah label data val:\n{df_val.jenis_luaran_lain.value_counts()}')
print(f'Jumlah label data test:\n{df_test.jenis_luaran_lain.value_counts()}')
# Status
print('=================================')
print(f'Jumlah label data asli:\n{df.status_luaran_lain.value_counts()}')
print(f'Jumlah label data train:\n{df_train.status_luaran_lain.value_counts()}')
print(f'Jumlah label data val:\n{df_val.status_luaran_lain.value_counts()}')
print(f'Jumlah label data test:\n{df_test.status_luaran_lain.value_counts()}')
```

Untuk beberapa pembagian data masih terdapat ketidaksama-rataan label, atau masih ada data yang belum memiliki label lain.

### Stratified Split

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 8:1:1
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['status_luaran_lain'])

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0, stratify=df_unseen['status_luaran_lain'])

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.status_luaran_lain.value_counts()}')
print(f'Jumlah label data train:\n{df_train.status_luaran_lain.value_counts()}')
print(f'Jumlah label data val:\n{df_val.status_luaran_lain.value_counts()}')
print(f'Jumlah label data test:\n{df_test.status_luaran_lain.value_counts()}')
```

penggunaan Stratified Split lebih efektif dari random split untuk pembagian label secara merata.

### K-Fold Cross Validation

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
df['tahun'] = label.fit_transform(df['tahun'])
df['skema'] = label.fit_transform(df['skema'])
df['jenis_luaran_hki'] = label.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = label.fit_transform(df['potensi_hki'])
df['jenis_luaran_lain'] = label.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = label.fit_transform(df['status_luaran_lain'])
df
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]

print(x)
print(y)
```

```python
from sklearn.model_selection import RepeatedKFold

# inisiasi repated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# inisiasi model
dt = DecisionTreeClassifier()
params = {'max_depth': list(range(5,11)),
          'criterion': ['gini', 'entropy', 'log_loss'],
          'splitter' : ['best', 'random']}
rs = RandomizedSearchCV(dt, params, cv=cv, n_jobs=-1)
%time rs.fit(x,y)
score = rs.score(x,y)
print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {rs.best_params_}')
```

```python
from sklearn.model_selection import GridSearchCV

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(dt, param_grid=params, cv=cv,n_jobs=-1)

# Fit / latih berdasarkan grid
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur
%timeit grid.fit(x, y)

# Evaluasi dengan score
score_gs = grid.score(x,y)

print(f'Hasil evaluasi: {score_gs}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

#### SVC

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
svc = SVC()
params_svc ={
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [1,5,10,50,100],
    'degree': [1,3,5,7,9]
}
rs_svc = RandomizedSearchCV(svc, params_svc, cv=cv, n_jobs=-1)
%time rs_svc.fit(x,y)
score_rs = rs_svc.score(x,y)
print(f'Hasil evaluasi: {score_rs}')
print(f'Konfigurasi hyperparameter: {rs_svc.best_params_}')
```

```python
from sklearn.model_selection import GridSearchCV

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid_gs = GridSearchCV(svc, param_grid=params_svc, cv=cv,n_jobs=-1)

# Fit / latih berdasarkan grid
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur
%timeit grid_gs.fit(x, y)

# Evaluasi dengan score
score_gs_svc = grid_gs.score(x,y)

print(f'Hasil evaluasi: {score_gs_svc}')
print(f'Konfigurasi hyperparameter: {grid_gs.best_params_}')
```
